```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
# from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

driver = webdriver.Chrome()
# Or for Firefox:
# driver = webdriver.Firefox(executable_path=driver_path)

# Open Discord
driver.get('https://discord.com')

# Remember to close the WebDriver when you're done
# driver.quit()
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, 'Login')))
driver.find_element(By.LINK_TEXT,"Login").click()

WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, 'email')))
driver.find_element(By.NAME, 'email').send_keys("mathpuli@gmail.com")

WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.NAME, 'password')))
driver.find_element(By.NAME, 'password').send_keys("dodgechallengerrt")

WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f'//div[contains(text(), "Log In")]')))
driver.find_element(By.XPATH, f'//div[contains(text(), "Log In")]').click()
```

```python
msg_elem = driver.find_element("xpath","//span[contains(@class, 'emptyText__13589')]")
msg_elem.send_keys("Hello world")
msg_elem.send_keys(Keys.RETURN)
```

## Hunt

```python
import random

for i in range (25):
    random_number = random.uniform(1, 2) * 3
    print(random_number)
    random_number+=15
    time.sleep(random_number)
    msg_elem = driver.find_element("xpath","//span[contains(@class, 'emptyText__13589')]")
    msg_elem.send_keys("oh")
    random_number = random.uniform(0, 1)
    time.sleep(random_number)
    msg_elem.send_keys(Keys.RETURN)
    print(">Number of Messages sent: "+str(i+1))
```
