```python
boss_element=["fire","light","earth","wind","metal","wood","ice","void"]
strong_list=[["fire", "earth"], ["wood", "light"], ["earth", "wind"],["wind", "ice"],["metal", "fire"],["wood", "fire"],["metal", "ice"],["void", "earth"]]
weak_list=[["metal","ice","wood"],["ice","void"],["metal","fire"],["wood","fire"],["earth","light"],["earth","light"],["wind","wood"],["metal","light"]]
```

```python
neutral_list=[]
for i in range(len(strong_list)):
    total=strong_list[i]+weak_list[i]
    temp=[]
    for j in boss_element:
        if j not in total:
            temp.append(j)
    neutral_list.append(temp)
print(neutral_list)
```

```python
import itertools
# Generate all possible combinations by taking one element from each sublist
all_possible_combinations = list(itertools.product(*strong_list))

min = 4

possible_decks=[]

for combination in all_possible_combinations:
    if len(list(set(combination)))==4:
        # print(set(combination))
        if set(combination) not in possible_decks:
            possible_decks.append(set(combination))


# print(result_list)
for i in possible_decks:
    print(i)
```
